```python
train_dir = path/"train"
valid_dir = path/"valid"
test_dir = path/"test"
train_json = path/"train"/"_annotations.coco.json"
valid_json = path/"valid"/"_annotations.coco.json"
test_json = path/"test"/"_annotations.coco.json"
```

```python
def load_df(json_path):
    with open(json_path, "r", encoding="utf-8") as f:
        coco = json.load(f)
    img = pd.DataFrame(coco["images"])
    anno = pd.DataFrame(coco["annotations"])
    merged_df = pd.merge(img, anno, left_on="id", right_on="image_id")
    df = merged_df[["file_name", "segmentation"]]
    return df
```

```python
train_df = load_df(train_json)
valid_df = load_df(valid_json)
test_df = load_df(test_json)
```

```python
import ast
class TumourDataset(Dataset):

    def __init__(self, df, img_dir):
        self.df = df.reset_index(drop=True)
        self.img_dir = img_dir

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        file_name = row["file_name"]

        polygon_coords = row["segmentation"]
        if isinstance(polygon_coords, str):
            try:
                polygon_coords = ast.literal_eval(polygon_coords)
            except (ValueError, SyntaxError):
                polygon_coords = []

        img_path = os.path.join(self.img_dir, file_name)
        img = cv.imread(img_path, cv.IMREAD_GRAYSCALE)

        if img is None:
            img = np.zeros((640, 640), dtype=np.uint8)

        mask = np.zeros((640, 640), dtype=np.uint8)

        if isinstance(polygon_coords, list) and len(polygon_coords) > 0:
            if isinstance(polygon_coords[0], list):
                flat_coords = polygon_coords[0]
            else:
                flat_coords = polygon_coords

            # Flatten any accidental deep nests and ensure even length for (x,y) pairs
            flat_coords = np.array(flat_coords).flatten()
            if len(flat_coords) % 2 == 0 and len(flat_coords) > 0:
                pts = np.array(flat_coords, dtype=np.int32).reshape(-1, 2)
                cv.fillPoly(mask, [pts], 255)

        # Normalization
        img_normalized = img / 255.0
        mask_binary = (mask / 255.0).astype(np.float32)

        img_tensor = torch.from_numpy(img_normalized).float().unsqueeze(0)
        mask_tensor = torch.from_numpy(mask_binary).float().unsqueeze(0)

        return img_tensor, mask_tensor
```

```python
import torch.nn as nn
class DoubleConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size = 3, bias=False, padding = 1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace = True ),
            nn.Conv2d(out_channels, out_channels, kernel_size = 3, bias=False, padding = 1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace = True)
        )
    def forward(self,x):
        return self.conv(x)
```

```python
class Down(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv = DoubleConv(in_channels, out_channels)
    def forward(self, x):
        x_pooled = self.pool(x)
        x_out = self.conv(x_pooled)
        return x_out
```

```python
class Up(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.up = nn.ConvTranspose2d(in_channels, in_channels//2, kernel_size = 2, stride = 2)
        self.conv = DoubleConv(in_channels, out_channels)
    def forward(self, x, skip_tensor):
        x_unsam = self.up(x)
        x_combined = torch.concat([x_unsam, skip_tensor],dim = 1)
        x_out = self.conv(x_combined)
        return x_out
```

```python
class CustomUNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.init_conv = DoubleConv(1,64)
        self.down1 = Down(64,128)
        self.down2 = Down(128,256)
        self.down3 = Down(256,512)
        self.up1 = Up(512,256)
        self.up2 = Up(256,128)
        self.up3 = Up(128,64)
        self.final_conv = nn.Conv2d(64,1,kernel_size = 1)
    def forward(self, x):
        skip1 = self.init_conv(x)
        skip2 = self.down1(skip1)
        skip3 = self.down2(skip2)
        bottleneck = self.down3(skip3)
        x_up = self.up1(bottleneck, skip3)
        x_up = self.up2(x_up, skip2)
        x_up = self.up3(x_up, skip1)
        out = self.final_conv(x_up)
        return out
```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import mlflow
import mlflow.pytorch
from app.predictor import NativeResNet34UNet
WEIGHTS_PATH = os.path.join("weights", "resnet34_brain_tumor.pth")
class HybridBCEDiceLoss(nn.Module):  
    def __init__(self, smooth=1.0):
        super().__init__()
        self.smooth = smooth
        self.bce = nn.BCEWithLogitsLoss()

    def forward(self, outputs, masks):
        bce_loss = self.bce(outputs, masks)  
        probs = torch.sigmoid(outputs)
        
        probs = probs.view(-1)
        masks = masks.view(-1)
        
        # Intersection calculation
        intersection = (probs * masks).sum()  # Hint: sum up the element-wise multiplication
        
        # Dice coefficient formula: (2 * intersection) / (total pixels predicted + total ground truth pixels)
        dice_coeff = (2. * intersection + self.smooth) / (probs.sum() + masks.sum() + self.smooth)
        dice_loss = 1.0 - dice_coeff
        
        return bce_loss + dice_loss
    
class BrainTumorDataset(Dataset):
    def __init__(self, num_samples=8):
        self.num_samples = num_samples

    def __len__(self):
        return self.num_samples

    def __getitem__(self, idx):
        # Generate a synthetic grayscale MRI scan slice [1, 640, 640]
        mock_image = torch.randn(1, 640, 640)
        # Generate a synthetic true mask slice [1, 640, 640] with zeros and ones
        mock_mask = (torch.rand(1, 640, 640) > 0.9).float()
        return mock_image, mock_mask


def run_training():
    epochs = 2
    batch_size = 2
    learning_rate = 0.001
    dataset = BrainTumorDataset(num_samples=4)
    train_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True) 
    
    model = NativeResNet34UNet()
    criterion = HybridBCEDiceLoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    
    mlflow.set_experiment("Brain_Tumor_Native_U-Net")
    
    print("--- Starting MLOps Training Experiment Run ---")
    with mlflow.start_run():
        # Log our structural configurations directly to the tracking server dashboard
        mlflow.log_param("epochs", epochs)
        mlflow.log_param("batch_size", batch_size)
        mlflow.log_param("learning_rate", learning_rate)
        
        for epoch in range(epochs):
            model.train()  # Explicitly activate training layers (BatchNorm/Dropout)
            running_loss = 0.0
            
            for batch_idx, (images, masks) in enumerate(train_loader):
                # Clear stale gradients from the previous computation step
                optimizer.zero_grad()
                
                # Forward Pass
                outputs = model(images)
                
                # Performance Evaluation
                loss = criterion(outputs, masks)
                
                # Backward Pass (Compute error gradients)
                loss.backward()
                
                # Weight Update
                optimizer.step()
                
                running_loss += loss.item()
            
            # Compute average performance metrics for the epoch
            epoch_loss = running_loss / len(train_loader)
            print(f"Epoch [{epoch+1}/{epochs}] - Computed Loss: {epoch_loss:.4f}")
            
            # Pipe metrics seamlessly into local MLflow telemetry
            mlflow.log_metric("loss", epoch_loss, step=epoch)
        
        # Post-Training Weight Preservation
        os.makedirs("weights", exist_ok=True)
        torch.save(model.state_dict(), WEIGHTS_PATH)
        print("--- Training Completed & Weights Saved Successfully ---")

if __name__ == "__main__":
    run_training()
```

```python
import torch.optim as optim
import segmentation_models_pytorch as smp
criterion = smp.losses.DiceLoss(mode='binary', from_logits=True)

#Pre-trained ResNet-34 U-Net
model_pretrained = smp.Unet(
    encoder_name="resnet34",
    encoder_weights="imagenet",
    in_channels=1,
    classes=1
).to(device)

#Custom Scratch-Built U-Net
model_scratch = CustomUNet().to(device)

optimizer_pretrained = optim.Adam(model_pretrained.parameters(), lr=1e-4)
optimizer_scratch    = optim.Adam(model_scratch.parameters(), lr=1e-4)
```

```python
import time

def train_one_epoch(model, dataloader, optimizer, criterion):
    model.train()
    total_loss = 0.0
    start_time = time.time()

    for batch_idx, (images, masks) in enumerate(dataloader):
        images = images.to(device)
        masks = masks.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, masks)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        if batch_idx % 40 == 0:
            print(f"  Batch {batch_idx}/{len(dataloader)} | Dice Loss: {loss.item():.4f}")

    epoch_time = time.time() - start_time
    average_loss = total_loss / len(dataloader)
    return average_loss, epoch_time
```

```python
NUM_EPOCHS = 10

pretrained_history = []
scratch_history = []

print("Training Pre-trained ResNet-34 U-Net")
for epoch in range(NUM_EPOCHS):
    print(f"🔄 Epoch {epoch+1}/{NUM_EPOCHS}")
    avg_loss, epoch_time = train_one_epoch(
        model_pretrained, train_load, optimizer_pretrained, criterion
    )
    pretrained_history.append(avg_loss)
    print(f"Avg Loss: {avg_loss:.4f} | Time: {epoch_time:.1f}s\n")

print("--------------------------------------------------\n")

print("Training Your Scratch-Built U-Net")
for epoch in range(NUM_EPOCHS):
    print(f"🔄 Epoch {epoch+1}/{NUM_EPOCHS}")
    avg_loss, epoch_time = train_one_epoch(
        model_scratch, train_load, optimizer_scratch, criterion
    )
    scratch_history.append(avg_loss)
    print(f"Avg Loss: {avg_loss:.4f} | Time: {epoch_time:.1f}s\n")
```

```python
import torch
import torch.nn.functional as F

def dice_loss(pred, target, smooth=1e-6):
    pred = torch.sigmoid(pred)  # Convert logits to probabilities
    intersection = (pred * target).sum(dim=(2, 3))
    union = pred.sum(dim=(2, 3)) + target.sum(dim=(2, 3))
    dice = (2. * intersection + smooth) / (union + smooth)
    return 1 - dice.mean()
```

```python
import torch.nn as nn
criterion_csn = nn.CrossEntropyLoss()
def train_one_epoch(model, dataloader, optimizer, criterion, criterion_csn):
    model.train()
    total_loss = 0.0
    start_time = time.time()

    for batch_idx, (images, masks) in enumerate(dataloader):
        images = images.to(device)
        masks = masks.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss_dice = criterion(outputs, masks)
        loss_ce = criterion_csn(outputs, masks)
        loss = loss_dice + loss_ce
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        if batch_idx % 40 == 0:
            print(f"  Batch {batch_idx}/{len(dataloader)} | Dice Loss: {loss.item():.4f}")

    epoch_time = time.time() - start_time
    average_loss = total_loss / len(dataloader)
    return average_loss, epoch_time
```

```python
NUM_EPOCHS = 5

pretrained_history = []
scratch_history = []

print("Training Pre-trained ResNet-34 U-Net")
for epoch in range(NUM_EPOCHS):
    print(f"🔄 Epoch {epoch+1}/{NUM_EPOCHS}")
    avg_loss, epoch_time = train_one_epoch(
        model_pretrained, train_load, optimizer_pretrained, criterion, criterion_csn
    )
    pretrained_history.append(avg_loss)
    print(f"Avg Loss: {avg_loss:.4f} | Time: {epoch_time:.1f}s\n")
```
